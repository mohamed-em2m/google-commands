```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras import Sequential
from tensorflow.keras.losses import MeanSquaredError, BinaryCrossentropy ,SparseCategoricalCrossentropy
from tensorflow.keras.activations import sigmoid
from tensorflow.keras.losses import MeanSquaredLogarithmicError
from tensorflow import keras	 as k	
import numpy as np
# for building linear regression models and preparing data
from sklearn.linear_model import LinearRegression ,LogisticRegression
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import tensorflow as tf
from xgboost import *
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler ,LabelEncoder
from tensorflow import keras
```

```python
x=[]
y=[]
num=0
```

data preprocessing withusiing mfcc

```python
for cat  in labels:
    print(cat)
    for aud in os.listdir(f"/content/{cat}/"):
        y.append(cat)
        red,sr=librosa.load(f"/content/{cat}/{aud}")
        print(len(red),num)
        
        red=librosa.resample(red,orig_sr=sr,target_sr=22050)
        red=np.concatenate((np.zeros(max(22050-len(red),0)),red))
        mfcc=librosa.feature.mfcc(y=red[:22050],n_mfcc=13).T
        x.append(mfcc)
        num+=1
```

```python
y2=np.array(y)
x2=np.array(x)
```

```python
encode=LabelEncoder()
y2[:]=encode.fit_transform(y2[:])
```

```python
pre=np.random.permutation(len(y2))
y2=y2[pre]
x2=x2[pre]
```

```python
y2=np.vectorize(int)(y2)
```

#first model

```python
model=keras.Sequential()
model.add(keras.layers.Flatten())
model.add(keras.layers.BatchNormalization())
#model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(256,activation=tf.nn.elu))
model.add(keras.layers.BatchNormalization())

model.add(keras.layers.Dropout(.3))
model.add(keras.layers.Dense(256,activation=tf.nn.elu))
model.add(keras.layers.BatchNormalization())

model.add(keras.layers.Dropout(.3))
model.add(keras.layers.Dense(256,activation=tf.nn.elu))
model.add(keras.layers.BatchNormalization())

model.add(keras.layers.Dropout(.3))
model.add(keras.layers.Dense(len(np.unique(y2)),activation='softmax'))
model.compile(loss=k.losses.SparseCategoricalCrossentropy(),optimizer='nadam', metrics=['accuracy'])
```

```python
model.fit(x2,y2,epochs=30,validation_split=.1,callbacks=k.callbacks.EarlyStopping(patience=10,restore_best_weights=True) )
```

# second model

```python
model3=keras.Sequential()
model3.add(    keras.layers.InputLayer(input_shape=(44,13)))
model3.add(keras.layers.BatchNormalization())
model3.add(keras.layers.LSTM(128 ,return_sequences=True))
model3.add(keras.layers.BatchNormalization())
model3.add(keras.layers.LSTM(128 ,return_sequences=True))
model3.add(keras.layers.BatchNormalization())
model3.add(keras.layers.LSTM(128 ,return_sequences=True))
model3.add(keras.layers.BatchNormalization())
model3.add(keras.layers.LSTM(128 ))
model3.add(keras.layers.BatchNormalization())
#model.add(keras.layers.Flatten())
model3.add(keras.layers.Dense(128,activation='elu'))
model3.add(keras.layers.BatchNormalization())

model3.add(keras.layers.Dense(len(np.unique(y2)),activation='softmax'))
model3.compile(loss=k.losses.SparseCategoricalCrossentropy(),optimizer='adam', metrics=['accuracy'])
```

```python
model3.fit(x2,y2,epochs=60,validation_split=.1,callbacks=k.callbacks.EarlyStopping(patience=10,restore_best_weights=True) )
```

# third model


```python
model6=keras.Sequential()
model6.add(    keras.layers.InputLayer(input_shape=(44,13)))
model6.add(keras.layers.BatchNormalization())
model6.add(keras.layers.LSTM(128 ))
model6.add(keras.layers.BatchNormalization())
model6.add(keras.layers.Flatten())
model6.add(keras.layers.BatchNormalization())
#model.add(keras.layers.Flatten())
model6.add(keras.layers.Dense(500,activation=tf.nn.elu))
model6.add(keras.layers.BatchNormalization())
model6.add(keras.layers.Dropout(.3))
model6.add(keras.layers.Dense(len(np.unique(y2)),activation='softmax'))
model6.compile(loss=k.losses.SparseCategoricalCrossentropy(),optimizer='nadam', metrics=['accuracy'])
```

```python
model6.fit(x2,y2,epochs=60,validation_split=.1,callbacks=k.callbacks.EarlyStopping(patience=10,restore_best_weights=True) )
```

# forth model 

```python
x2=x2.reshape((105829, 44*  13,1))

model1=keras.Sequential()
model1.add(    keras.layers.InputLayer(input_shape=(44* 13,1)))
model1.add(keras.layers.BatchNormalization())
model1.add(keras.layers.Conv1D(filters=32, kernel_size=5, activation='relu'))
model1.add(keras.layers.MaxPooling1D(pool_size=2))

model1.add(keras.layers.BatchNormalization())

model1.add(keras.layers.MaxPooling1D(pool_size=2))
model1.add(keras.layers.Conv1D(filters=64, kernel_size=5, activation='relu'))
model1.add(keras.layers.BatchNormalization())

model1.add(keras.layers.MaxPooling1D(pool_size=2))
model1.add(keras.layers.Flatten())

#model.add(keras.layers.Flatten())
model1.add(keras.layers.Dense(256,activation=tf.nn.elu))
model1.add(keras.layers.BatchNormalization())

model1.add(keras.layers.Dropout(.3))
model1.add(keras.layers.Dense(128,activation=tf.nn.elu))
model1.add(keras.layers.BatchNormalization())

model1.add(keras.layers.Dropout(.3))
model1.add(keras.layers.Dense(len(np.unique(y2)),activation='softmax'))
model1.compile(loss=k.losses.SparseCategoricalCrossentropy(),optimizer='nadam', metrics=['accuracy'])
```

```python
x2=x2.reshape((105829, 44*  13,1))

model1.fit(x2,y2,epochs=30,validation_split=.1,callbacks=k.callbacks.EarlyStopping(patience=10,restore_best_weights=True) )
```

# sixth model

```python
x2=x2.reshape((105829, 44, 13,1))
model2=keras.Sequential()
model2.add(keras.layers.BatchNormalization())

model2.add(keras.layers.Conv2D(filters=32, kernel_size=(3,3), activation='relu'))
model2.add(keras.layers.BatchNormalization())
model2.add(keras.layers.Conv2D(filters=64, kernel_size=(3,3), activation='relu'))
model2.add(keras.layers.BatchNormalization())

model2.add(keras.layers.MaxPooling2D(pool_size=(2,2)))
model2.add(keras.layers.Conv2D(filters=128, kernel_size=(3,3), activation='relu'))
model2.add(keras.layers.BatchNormalization())

model2.add(keras.layers.MaxPooling2D(pool_size=(2,2)))
model2.add(keras.layers.Flatten())

#model.add(keras.layers.Flatten())
model2.add(keras.layers.Dense(256,activation=tf.nn.elu))
model2.add(keras.layers.BatchNormalization())

model2.add(keras.layers.Dropout(.3))
model2.add(keras.layers.Dense(256,activation=tf.nn.elu))
model2.add(keras.layers.BatchNormalization())

model2.add(keras.layers.Dropout(.3))
model2.add(keras.layers.Dense(len(np.unique(y2)),activation='softmax'))
model2.compile(loss=k.losses.SparseCategoricalCrossentropy(),optimizer='adam', metrics=['accuracy'])
```
